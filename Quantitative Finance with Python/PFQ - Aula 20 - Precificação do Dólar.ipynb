```python
import numpy as np
import matplotlib.pyplot as plt

# define the starting price of BRL=x
start_price = 5.6

# define the number of simulations to run
num_sims = 10000

# define the volatility (i.e. the standard deviation of daily returns)
volatility = 0.1

# define the time horizon (i.e. the number of days to simulate)
time_horizon = 20

# define the daily returns
daily_returns = np.random.normal(loc=0.0, scale=volatility, size=(time_horizon, num_sims))

# calculate the price of BRL=x for each day
prices = np.zeros_like(daily_returns)
prices[0] = start_price
for i in range(1, time_horizon):
    prices[i] = prices[i-1] * (1 + daily_returns[i])

# calculate the average final price of BRL=x
avg_price = np.mean(prices[-1])

avg_prices = np.mean(prices, axis = 1)

# calculate the confidence interval for the final price of BRL=x
conf_int = np.percentile(prices[-1], [5, 95])

# plot the prices over time
plt.plot(avg_prices)
plt.xlabel('Time (days)')
plt.ylabel('Price of BRL=x')
plt.show()

# print the results
print("Average final price of BRL=x: ", avg_price)
print("Confidence interval for the final price of BRL=x: ", conf_int)
```
